```python
R0_v1 = venezuela / 28328617
I0_v1 = 0.000006 - R0_v1
S0_v1 = 1 - I0_v1 - R0_v1
print ("S0: ", S0_v1)
print ("I0: ", I0_v1)
print ("R0: ", R0_v1)
```

```python
def equations(t, y):
    i = 1 -y[0]-y[1]
    
    dSdt = -a_v1 * y[0] * i
    dRdt = b * i
    
    return [dSdt, dRdt]

from scipy.integrate import solve_ivp
day = np.asarray(range(1,Cumulative_cases_3C['Venezuela_CC'].shape[0]+1))
t = np.linspace(0, 180,527)
x_initial = S0_v1, R0_v1
solnv1 = solve_ivp(equations, [1, Cumulative_cases_3C['Venezuela_CC'].shape[0]], x_initial, t_eval=day)
from sklearn.metrics import mean_squared_error
pv1 = 1 - solnv1.y[0, :]
MSE_v1 = np.mean(np.square(pv1 - Cumulative_cases_3C['Venezuela_CC']))
print(a_v1,MSE_v1)
```
